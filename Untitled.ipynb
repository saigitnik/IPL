```python
df=match['toss_decision'].value_counts(normalize=True)*100
```

```python
plt.figure(figsize=(8, 6))
df.plot(kind='bar', color='blue')
plt.title('Frequency of Toss Decisions')
plt.xlabel('Toss Decision')
plt.ylabel('Frequency')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
df.plot(kind='pie', autopct='%1.1f%%', colors=['lightblue', 'green'])
plt.title('Distribution of Toss Decisions')
plt.ylabel('')
plt.show()
```
